```python
import numpy as np
import matplotlib.pyplot as plt
# Assuming your cleaned and processed dataset is stored in a DataFrame called 'df'

# Define the possible values for 'Location' and 'device_type'
locations = ['region1', 'region2', 'region3', 'region4', 'region5']
device_types = ['router', 'hub', 'switches', 'server']

# Add the 'Location' column with random values
df['Location'] = np.random.choice(locations, size=len(df))

# Add the 'device_type' column with random values
df['device_type'] = np.random.choice(device_types, size=len(df))

# Display the DataFrame with the new columns
print(df)
```

```python
original_shape = df.shape

# Reduce the size of the dataset by randomly sampling 50% of the data
df_sampled = df.sample(frac=0.5, random_state=1)

# Display the shape of the sampled dataset
sampled_shape = df_sampled.shape

original_shape, sampled_shape
```

```python
nan_values = df_sampled.isna().sum()

print("NaN values in the dataset:")
print(nan_values)
```

```python
df_sampled['protocol_type'] = df_sampled['protocol_type'].astype('category').cat.codes

df_sampled['service'] = df_sampled['service'].astype('category').cat.codes

df_sampled['connection_status'] = df_sampled['connection_status'].astype('category').cat.codes

# Display the head of the dataframe after converting categorical features
df_sampled
```

```python
df_sampled['attack_type'] = df_sampled['attack_type'].astype('category').cat.codes

df_sampled['Location'] = df_sampled['Location'].astype('category').cat.codes

df_sampled['device_type'] = df_sampled['device_type'].astype('category').cat.codes

# Display the head of the dataframe after converting categorical features
df_sampled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, confusion_matrix

# Split the data into features and target
X = df_sampled.drop(columns=['attack_type'])
y = df_sampled['attack_type']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the SVM model
svm_model = SVC()

# Fit the model on the training data
svm_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svm_model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("accuracy:",accuracy)
# precision
precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", precision)
#recall
recall = recall_score(y_test, y_pred, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("Recall:", recall)

# Calculate the F1 score of the model
f1 = f1_score(y_test, y_pred, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("F1 Score:", f1)

# Calculate the confusion matrix of the model
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import time

# Define a function to train and test a model and measure execution time
def train_test_model(model, X_train, y_train, X_test, y_test):
    # Start the timer
    start_time = time.time()

    # Train the model
    model.fit(X_train, y_train)

    # Measure the training time
    train_time = time.time() - start_time

    # Start the timer for testing
    start_time = time.time()

    # Test the model
    y_pred = model.predict(X_test)

    # Measure the testing time
    test_time = time.time() - start_time

    # Return the predictions and execution times
    return y_pred, train_time, test_time

# Example usage:
# Assuming model is your machine learning model, X_train and y_train are your training data, and X_test and y_test are your test data
# Replace model, X_train, y_train, X_test, and y_test with your actual data and model

y_pred, train_time, test_time = train_test_model(svm_model, X_train, y_train, X_test, y_test)
print("Training time:", train_time)
print("Testing time:", test_time)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the Decision Tree model
dt_model = DecisionTreeClassifier()

# Fit the Decision Tree model on the training data
dt_model.fit(X_train, y_train)

# Make predictions on the test data using Decision Tree
y_pred_dt = dt_model.predict(X_test)

# Calculate the accuracy of the Decision Tree model
accuracy_dt = accuracy_score(y_test, y_pred_dt)

# Initialize the Random Forest model
rf_model = RandomForestClassifier()

# Fit the Random Forest model on the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test data using Random Forest
y_pred_rf = rf_model.predict(X_test)

# Calculate the accuracy of the Random Forest model
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print("accuracy for decision tree:",accuracy_dt)
print("accuracy for random forest",accuracy_rf)
# precision
precision = precision_score(y_test, y_pred_dt, average='weighted')
print("Precision for decision tree:", precision)
precision = precision_score(y_test, y_pred_rf, average='weighted')
print("Precision for random forrest:", precision)
#recall
recall = recall_score(y_test, y_pred_dt, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("Recall for decision tree:", recall)
recall = recall_score(y_test, y_pred_rf, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("Recall random forrest:", recall)

# Calculate the F1 score of the model
f1 = f1_score(y_test, y_pred_dt, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("F1 Score for decision tree:", f1)
f1 = f1_score(y_test, y_pred_rf, average='weighted')  # or 'macro', 'micro', 'binary' depending on your requirement
print("F1 Score for random forrest:", f1)

# Calculate the confusion matrix of the model
conf_matrix = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix for decision tree:")
print(conf_matrix)
conf_matrix = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix for random forrest:")
print(conf_matrix)
```
